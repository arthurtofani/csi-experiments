```python
import matplotlib.pyplot as plt
from csi import approach_a as exp
```

```python
def load_audios(audio_files):
    arr = []
    for audio_file in audio_files:
        print('====Loading==== %s' % audio_file)
        audio = exp.Audio(audio_file)
        audio.extract_chroma(hop_length=2**11)
        audio.extract_ssm()
        audio.extract_segments(size=64, max_levels=6)    
        arr.append(audio)
    return arr
        
def plot_audio(aud):
    plt.figure(figsize=(14, 7))
    plt.subplot(1, 2, 1)
    aud.plot_ssm()
    aud.plot_profile()
    plt.subplot(1, 2, 2)
    aud.plot_ssm()
    aud.plot_segments()
    plt.show()

def plot_peaks(audios):
    plt.figure(figsize=(12, 10))
    idx=0
    for audio in audios:
        idx+=1
        plt.subplot(len(audios), 1, idx)
        audio.plot_ssm()
        audio.plot_ssm()
        audio.plot_peaks()
        plt.title(audio.filename)
    plt.show()
```

```python
letitbe_files = [
    '/dataset/YTCdataset/letitbe/v0.mp3',
    '/dataset/YTCdataset/letitbe/v1.mp3',
    '/dataset/YTCdataset/letitbe/v2.mp3',
    '/dataset/YTCdataset/letitbe/v3.mp3'
]
letitbe = load_audios(letitbe_files)
```

```python
jbg_files = [
    '/dataset/YTCdataset/jbg/v0.mp3',
    '/dataset/YTCdataset/jbg/v1.mp3',
    '/dataset/YTCdataset/jbg/v2.mp3',
    '/dataset/YTCdataset/jbg/v3.mp3'
]
jbg = load_audios(jbg_files)
```

```python
plot_peaks(letitbe)
```
